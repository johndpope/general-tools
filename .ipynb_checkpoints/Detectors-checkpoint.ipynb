# Detectors
Trying to train a "detector" - a single conv filter that should learn a pattern that appears in all images

```python
import keras
from keras.models import Sequential
from keras.layers import *
from keras.regularizers import *
from keras.optimizers import *
from keras.datasets import cifar10
import keras.backend as K
import time
import math
import scipy

import matplotlib
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec

import sklearn.model_selection

%matplotlib inline

import ipywidgets as widgets
import IPython

# why? https://github.com/fchollet/keras/issues/3959
#K.set_image_dim_ordering("th")
K.set_image_dim_ordering("tf")
```

```python
def visualize_grid(Xs, ubound=255.0, padding=1):
  """
  Reshape a 4D tensor of image data to a grid for easy visualization.

  Inputs:
  - Xs: Data of shape (N, H, W, C)
  - ubound: Output grid will have values scaled to the range [0, ubound]
  - padding: The number of blank pixels between elements of the grid
  """
  (N, H, W, C) = Xs.shape
  grid_size = int(math.ceil(math.sqrt(N)))
  grid_height = H * grid_size + padding * (grid_size - 1)
  grid_width = W * grid_size + padding * (grid_size - 1)
  grid = np.zeros((grid_height, grid_width, C))
  next_idx = 0
  y0, y1 = 0, H
  for y in range(grid_size):
    x0, x1 = 0, W
    for x in range(grid_size):
      if next_idx < N:
        img = Xs[next_idx]
        low, high = np.min(img), np.max(img)
        #grid[y0:y1, x0:x1] = ubound * (img - low) / (high - low)
        grid[y0:y1, x0:x1] = Xs[next_idx]
        next_idx += 1
      x0 += W + padding
      x1 += W + padding
    y0 += H + padding
    y1 += H + padding
  # grid_max = np.max(grid)
  # grid_min = np.min(grid)
  # grid = ubound * (grid - grid_min) / (grid_max - grid_min)
  return grid
```

## Generate images
Generate N images, starting from some noise (0 or 1) and adding the cross in random places

```python
N = 1000 # num images
W = 40 # width
H = 40 # height

DIMS = 3
CROSS_W, CROSS_H = (5,5)

def cross_img(h,w):
    a = np.zeros((h,w, DIMS))
    if w % 2 == 0:
        a[:, int(w/2)-1:int(w/2)+1, :] = 1
    else:
        a[:, w//2:w//2+1, :] = 1
    if h % 2 == 0:
        a[int(h/2)-1:int(h/2)+1, :, :] = 1
    else:
        a[h//2:h//2+1, :, :] = 1
    return a

cross = cross_img(CROSS_H, CROSS_W)
images = np.random.random((N, H, W, DIMS))


# place cross randomly
for i in range(N//2):
    rh = np.random.randint((H-CROSS_H+1)//2)
    rw = np.random.randint((W-CROSS_W+1)//2)
    images[i, rh:rh+CROSS_H, rw:rw+CROSS_W][cross==1] = 1

ys = np.append(np.ones((N//2,)), np.zeros((N-N//2,)))
#ys_oh = keras.utils.np_utils.to_categorical(ys)

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(images, ys, test_size=0.1)
y_train_oh = keras.utils.np_utils.to_categorical(y_train)
y_test_oh = keras.utils.np_utils.to_categorical(y_test)


grid = visualize_grid(images).squeeze()

plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(grid, cmap=plt.cm.gray, interpolation='nearest')

```

```python
X_train_means = X_train.mean(axis=0)
X_train_std = X_train.std(axis=0)
X_train = (X_train - X_train_means) / X_train_std
X_test = (X_test - X_train_means) / X_train_std
```

# Train the network
Input -> Conv2D(1, size=(FH,FW)) -> MaxPooling(all image) -> (1,) activation

```python
def create_model(opt='adam', num_filters=(16,4), gradual_pooling=False, batchnorm=True):
    model = Sequential()

    # Adding CONV0 layers
    conv0_act = None if batchnorm else 'relu'
    model.add(Convolution2D(num_filters[0],3,3, name='conv0_0', activation='relu', input_shape=(H,W,DIMS)))
    model.add(Convolution2D(num_filters[1],3,3, name='conv0_1', activation=conv0_act))
    if batchnorm:
        model.add(BatchNormalization(name='bn_conv0', momentum=0.9, mode=0, axis=-1))
        model.add(Activation('relu'))
        

    # Adding MAXPOOLING 
    if gradual_pooling:
        for i in range(int(math.log2(H)-1)):
            model.add(MaxPooling2D((2,2), name="maxpooling_conv0_{}".format(i)))
    ll = model.layers[-1].output_shape[1:3]
    model.add(MaxPooling2D(ll, name='maxpooling_conv0'))

    # FC to outputs
    model.add(Flatten())
    model.add(Dense(2, name='output', activation='softmax'))
#    model.add(Activation('softmax'))
    
    # Compile model
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

    return model          
```

# Analyzing layers (post training) - Activation maps, representative inputs

```python
def get_layer_desc(layer):
    if isinstance(layer, keras.layers.Flatten):
        return ""
    elif isinstance(layer, keras.layers.Dense):
        return ""
    elif isinstance(layer, keras.layers.Convolution2D):
        return "{}x({},{})".format(layer.nb_filter, layer.nb_row, layer.nb_col)
    elif isinstance(layer, keras.layers.MaxPooling2D):
        return layer.pool_size
    elif isinstance(layer, keras.layers.normalization.BatchNormalization):
        return "BatchNormalization"
        

def analyze_layer(model, layer_num, trainingset=None, calculate_repr_inputs=True):
    model_input_shape = model.input_shape[1:]
    layer = model.layers[layer_num]
    layer_name = type(layer).__name__
    print("analyzing layer {}...".format(layer_name))
    
    l_input_shape = layer.input.get_shape()
    l_output_shape = layer.output.get_shape()
    
    res = dict( 
        name=layer_name, 
        desc=get_layer_desc(layer),
        input_shape=l_input_shape,
        output_shape=l_output_shape
    )
    
    w = layer.get_weights()
    if type(layer) in [keras.layers.Dense, keras.layers.Convolution2D]: 
        l_w, l_b = w
        num_elements = l_b.shape[0]
        element_shape = l_w.shape[:len(l_w.shape)-1]

        res['num_learning_elements'] = num_elements
        res['learning_element_shape'] = element_shape
        
        input_img = model.layers[0].input
    
        # Calculating the representative inputs
        repr_inputs = np.zeros((num_elements, ) + model_input_shape)
        if calculate_repr_inputs:
            for element_id in range(num_elements):
                # NOTICE: repr element is neuron (0,0) in the conv act map. this is also used when
                # calculating avg neuron affect on output
                repr_element_id = [0]*(len(l_output_shape)-1) + [element_id]

                filter_act = layer.output[repr_element_id] - K.sum(input_img)
                grads = K.gradients(filter_act, input_img)[0] 
                grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5)
                it = K.function([input_img, K.learning_phase()], [filter_act, grads])

                learning_rate = 1
                num_iters = 100
                input_shape = (1,)+model_input_shape
                curr_input = np.zeros(input_shape)
                if it([curr_input, False])[0] == 0:
                    curr_input = np.random.random(input_shape)

                for i in range(num_iters):
                    curr_act, curr_grads = it([curr_input, False])
                    curr_input += learning_rate * curr_grads
                    learning_rate *= 0.98

                repr_inputs[element_id,:] = curr_input

            res['learning_elements_repr_inputs'] = repr_inputs

        if trainingset is not None:
#             # Calculating average neuron effect on the final output
#             d_output_layer_output_func = K.function([model.layers[0].input, K.learning_phase()], [K.gradients(model.output, layer.output)[0]])
#             d_output_layer_output = d_output_layer_output_func([trainingset, False])[0]
#             res['grads_on_output'] = d_output_layer_output.mean(axis=tuple(range(len(l_output_shape)-1)))

            # Average Activation Map
            act_map = K.function([model.layers[0].input, K.learning_phase()], [layer.output])
            res['activation_map'] = act_map([trainingset, False])[0].mean(axis=0)
            
        
    return res
```

```python
def display_analyzed_layer(anl_layer, display_gray=False, display_limit=-1):
    print("{} {}".format(anl_layer['name'], anl_layer['desc']))
    print("Input: {}".format(anl_layer['input_shape']))
    print("Output: {}".format(anl_layer['output_shape']))


    if 'num_learning_elements' in anl_layer:

        num_ri = anl_layer['num_learning_elements']

        plt.figure(figsize=(25,10))
        gs = gridspec.GridSpec(int(math.ceil(num_ri/8)), 8)
        gs.update(wspace=0.025, hspace=0.05)

        inner_gs = [ gridspec.GridSpecFromSubplotSpec(2, 1, subplot_spec=gs[i]) for i in range(num_ri) ]

        for i in range(num_ri):
            goo = "{:.2e}".format(anl_layer['grads_on_output'][i]) if 'grads_on_output' in anl_layer else 'N/A'

            if 'learning_elements_repr_inputs' in anl_layer:
                repr_inputs = anl_layer['learning_elements_repr_inputs']
                inp = repr_inputs[i,:].squeeze()
                inp = (inp * X_train_std) + X_train_means

                ri_ax = plt.subplot(inner_gs[i][0])
                ri_ax.set_title("Repr. Input #{}\nGrad on output {}".format(i, goo))
                ri_ax.set_xticklabels([])
                ri_ax.set_yticklabels([])
                ri_ax.set_aspect('auto')        
                
                img = inp[:display_limit,:display_limit]
                if display_gray:
                    img = img.mean(axis=2)

                ri_ax.imshow(img, cmap=plt.cm.gray, interpolation='nearest')

            if 'activation_map' in anl_layer:
                actmap = np.moveaxis(anl_layer['activation_map'],-1,0)[i]
                if len(actmap.shape) == 0: continue
                
                act_ax = plt.subplot(inner_gs[i][1])
                act_ax.set_xticklabels([])
                act_ax.set_yticklabels([])
                act_ax.set_aspect('auto')        
                act_ax.imshow(actmap, cmap=plt.cm.gray, interpolation='nearest')
                act_ax.set_title("Activation Map #{}\nGrad on output {}".format(i,goo))
        plt.show()

    

```

# Plotting Dead Neurons

```python
class TrackDeadNeurons(keras.callbacks.Callback):
    def __init__(self):
        pass
    
    def on_train_begin(self, logs={}):
        self.dead_neurons = [ [] for l in self.model.layers ]
        self.actmap_func = K.function( [self.model.layers[0].input, K.learning_phase()], [l.output for l in self.model.layers])
        
    # batch - batch num (int)
    # self.model - keras.models.Sequential
    def on_epoch_end(self, batch, logs={}):
        layer_actmaps = self.actmap_func([self.model.validation_data[0], False])
        for l_idx, actmap in enumerate(layer_actmaps):
            output_component_deads = ((actmap != 0).sum(axis=tuple(range(len(actmap.shape)-1))) == 0)
            pct_dead = output_component_deads.sum() / len(output_component_deads)
            self.dead_neurons[l_idx].append(pct_dead)

    
test_model = create_model(num_filters=(16,16), batchnorm=False)

track_dead_neurons = TrackDeadNeurons()
test_model.fit(X_train, y_train, nb_epoch=50, verbose=2, validation_data=(X_train, y_train),
                callbacks=[
                    track_dead_neurons,
                ])

```

# Occluding parts of the image

```python
num_imgs = 6
imgs = X_train[:num_imgs]

def occluding_analysis(img, patch=(3,3)):
    res = np.zeros(img.shape[:2])
    for i in range(0,img.shape[0],patch[0]):
        for j in range(0,img.shape[1],patch[1]):
            occ_img = img.copy()
            occ_img[i:i+patch[0], j:j+patch[1], :] = 0
            res[i:i+patch[0], j:j+patch[1]] = model.predict(occ_img[None,:])
    return res

ress = [ occluding_analysis(img, patch=(5,5)) for img in imgs ]

fig = plt.figure(figsize=(20,10))
gs = gridspec.GridSpec(2,6)

for idx, (img, res) in enumerate(zip(imgs, ress)):
    ax = plt.Subplot(fig, gs[0, idx])
    ax.imshow(img, interpolation='nearest')
    fig.add_subplot(ax)

    ax = plt.Subplot(fig, gs[1, idx])
    ax.imshow(res, cmap=plt.cm.jet_r, vmin=0, vmax=1)
    fig.add_subplot(ax)

```
